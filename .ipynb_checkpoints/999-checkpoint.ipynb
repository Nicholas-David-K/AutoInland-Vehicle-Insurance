#### Convert to appropriate data types

```python
# dates
date_columns = [col for col in df_combined.columns if "Date" in col]
for col in date_columns:
    df_combined[col] = pd.to_datetime(df_combined[col])
```

#### Fill in missing values

```python
df_combined[['LGA_Name','State']]
```

```python
for key, row in df_combined.iterrows():
    if(row.LGA_Name == 'Victoria Island'):
        df_combined['State'].fillna('Lagos', inplace=True)
```

```python
df_combined[['LGA_Name', 'State']]
```

```python
# Combine LGA name and State
from collections import OrderedDict
import re

city_and_state = df_combined['State'] + " " + df_combined['LGA_Name']
city_and_state = city_and_state.apply(lambda x: re.sub('[*=/,&^#@]+', ' ', x))
city_and_state = city_and_state.str.replace('-', ' ')
df_combined['City_State'] = city_and_state.str.split().apply(lambda x: OrderedDict.fromkeys(x).keys()).str.join(" ")

# drop lga and state
df_combined.drop(['LGA_Name', 'State'], axis=1, inplace=True)
```

```python
# Gender 
'''
We will have categories Male, Female, and Other. Let's get on with it.
'''
mapper = {
    'Male': 'Male',
    'Female': 'Female',
    'Entity': 'Other',
    'Joint Gender': 'Other',
    'NOT STATED': 'Other',
    'NO GENDER': 'Other',
    np.nan : np.nan,
    'SEX': 'Other'
}

df_combined['Gender'] = df_combined['Gender'].apply(lambda x: mapper[x])

'''
I will then replace the missing values with the most frequent Gender which is Male
'''
df_combined['Gender'].fillna('Male', inplace=True)
```

```python
df_combined['Gender'].value_counts()
```

```python
# Product Name
'''
Most Occurring Product Name is Car Classic. Replace all missing values with Car Classic
'''
df_combined['ProductName'].fillna('Car Classic', inplace=True)
```

```python
claims = df_combined['target'].value_counts()
'''
Most People did not submit a claim. So I will replace missing target values with 0
'''
print('Number of claims NOT  submitted is {} while number of claims submitted is {}.'.format(*df_combined['target'].value_counts()))

# replace missing with 0
df_combined['target'].fillna(0, inplace=True)
```

```python
# Check missing values
df_combined.isnull().sum() # Hoorah! No missing values
```

#### Feature Engineering
- Extract information from other columns in the dataset

```python
date_columns = ['Policy Start Date', 'Policy End Date', 'First Transaction Date']
date_extractions = ['year', 'dayofyear', 'month', 
                    'days_in_month', 'day', 'dayofweek', 
                    'weekofyear', 'quarter',  'week', 
                    'is_quarter_start', 'is_quarter_end', 'is_month_start', 
                    'is_month_end', 'is_year_start', 'is_year_end']
for col in date_columns:
    for feature in date_extractions:
        df_combined[col+"_"+feature] = getattr(df_combined[col].dt, feature)

        
# get weekend 0, 1 (1=weekend, 0=not weekend)
for col in date_columns:
    df_combined[col+"_"+'weekend'] = np.where((df_combined[col].dt.day_name()).isin(['Sunday','Saturday']), 1, 0)
```

```python
le = ce.OneHotEncoder()
date_bools = [col for col in df_combined if "is_" in col]
for date_b in date_bools:
    df_combined[date_b] = le.fit_transform(df_combined[date_b])
```

## **Outliers**

```python
def outliers(column):
    Q1, Q3 = np.percentile(column, [25, 75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range, upper_range
 
outliers(df_combined['Age'])
```

```python
age_mean = df_combined[(df_combined['Age'] > 12.5) & (df_combined['Age'] < 72.5)]['Age'].mean()
age_mean
```

```python
df_combined['Age'] = np.where(df_combined['Age'] < 12.5, age_mean, df_combined['Age'])
```

```python
df_combined['Age'] = np.where(df_combined['Age'] > 72.5, age_mean, df_combined['Age'])
```

## **Init Model**

```python
def train_predict(classifier, X_train, y_train, X_test, y_test):
    ''''''
    results = {}
    
    # Training
    start = time()
    clf = classifier.fit(X_train, y_train)
    end = time()
    
    results['train_time'] = end - start
    
    # Testing
    start = time()
    predictions = clf.predict(X_test)
    end = time()
    
    results['test_time'] = end - start
    
    # Accuracy score
#     score = accuracy_score(y_test, predictions)
#     results['accuracy_score'] = score
    
    # f1_score
    f_score = f1_score(y_test, predictions)
    results['f1_score'] = f_score
    
#     # cross val-score
#     cross_scores = cross_val_score(clf, X_train, y_train, scoring='f1', cv=5)
#     results['f1_validation_score'] = cross_scores.mean()
    
    print('{} has been trained and tested \U0001F600'.format(classifier.__class__.__name__))
    return results
```

```python
# Separate training set and test test
training_df = df_combined[:training_data]
testing_df = df_combined[training_data:]

main_cols = [col for col in df_combined[training_data:].columns if 'Target' not in col] # remove the Target variable from the training set
testing_df = testing_df[main_cols]

# training and testing shapes (Original Format)
training_df.shape, testing_df.shape
```

```python
# correlation between test harness and ideal test condition
from numpy import mean
from numpy import isnan
from numpy import asarray
from numpy import polyfit
from scipy.stats import pearsonr
from matplotlib import pyplot
from sklearn.datasets import make_classification
from sklearn.model_selection import KFold
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import RidgeClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import ExtraTreeClassifier
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

def get_models():
    models = list()
    models.append(LogisticRegression())
    models.append(RidgeClassifier())
    models.append(SGDClassifier())
#     models.append(PassiveAggressiveClassifier())
#     models.append(KNeighborsClassifier())
#     models.append(DecisionTreeClassifier())
#     models.append(ExtraTreeClassifier())
#     models.append(LinearSVC())
#     models.append(SVC())
#     models.append(GaussianNB())
#     models.append(AdaBoostClassifier())
#     models.append(BaggingClassifier())
#     models.append(RandomForestClassifier())
#     models.append(ExtraTreesClassifier())
#     models.append(GaussianProcessClassifier())
#     models.append(GradientBoostingClassifier())
#     models.append(LinearDiscriminantAnalysis())
#     models.append(QuadraticDiscriminantAnalysis())
#     models.append(LGBMClassifier())
#     models.append(MLPClassifier())
    return models

def get_dataset():
    X, y = make_classification(n_samples=100, n_features=20, n_informative=5, n_redundant=5, random_state=1)
    return X, y

def evaluate_model(cv, model):
    X_train_val, y_train_val = X_train, y_train
    scores = cross_val_score(model, X_train_val, y_train_val, scoring='f1', cv=cv, n_jobs=-1)
    return mean(scores)


ideal_cv = LeaveOneOut()
cv = KFold(n_splits=5, shuffle=True, random_state=1)
models = get_models()
ideal_results, cv_results = list(), list()

for model in models:
    cv_mean = evaluate_model(cv, model)
    ideal_mean = evaluate_model(ideal_cv, model)
    if isnan(cv_mean) or isnan(ideal_mean):
        continue
    cv_results.append(cv_mean)
    ideal_results.append(ideal_mean)
    print('>%s: ideal=%.3f, cv=%.3f' % (type(model).__name__, ideal_mean, cv_mean))
    
# calculate the correlation between each test condition
corr, _ = pearsonr(cv_results, ideal_results)
print('Correlation: %.3f' % corr)

# scatter plot of results
pyplot.scatter(cv_results, ideal_results)

# plot the line of best fit
coeff, bias = polyfit(cv_results, ideal_results, 1)
line = coeff * asarray(cv_results) + bias
pyplot.plot(cv_results, line, color='r')

# label the plot
pyplot.title('10-fold CV vs LOOCV Mean Accuracy')
pyplot.xlabel('Mean Accuracy (10-fold CV)')
pyplot.ylabel('Mean Accuracy (LOOCV)')
# show the plot
pyplot.show()
```
