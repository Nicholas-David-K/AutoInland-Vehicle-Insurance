#### Convert to appropriate data types

```python
# dates
date_columns = [col for col in df_combined.columns if "Date" in col]
for col in date_columns:
    df_combined[col] = pd.to_datetime(df_combined[col])
```

#### Fill in missing values

```python
# Subject_Car_Colour
color_freq = df_combined['Subject_Car_Colour'].value_counts().nlargest(1)
print("The most popular color is Black with {} values recorded.".format(color_freq[0]))

# replace missing values in Subject_Car_Colour with color Black
df_combined['Subject_Car_Colour'].fillna('Black', inplace=True)
```

```python
# Car_Category
car_cat_freq = df_combined['Car_Category'].value_counts().nlargest(1)
print("The most popular car category is Saloon with {} values recorded.".format(car_cat_freq[0]))

# replace missing values in Subject_Car_Colour with color Black
df_combined['Car_Category'].fillna('Saloon', inplace=True)
```

```python
# Subject_Car_Make
subject_car_freq = df_combined['Subject_Car_Make'].value_counts().nlargest(1)
print("Irrespective of the Car_Category, the most occurring Subject_Car_Make is TOYOTA with {} values recorded.".format(subject_car_freq[0]))

# replace missing values with TOYOTA
df_combined['Subject_Car_Make'].fillna("TOYOTA", inplace=True)
```

```python
# LGA_Name
place_freq = df_combined['LGA_Name'].value_counts().nlargest(1)
print("Most Policies were purchased at Victoria Island with {} values recored.".format(place_freq[0]))

# replace missing values with TOYOTA
df_combined['LGA_Name'].fillna("Victoria Island", inplace=True)
```

```python
'''
Missing values in State are in LGA_Name "Victoria Island" (City where Policy was purchased). State with no missing values with Victoria Island as the purchasing state
are in State "Lagos". The only occurrence where the City is not Lagos is the record printed below. I will replace missing values in State column with value "Lagos" including the record with
State Name 'N/A'
'''
df_combined[(df_combined['LGA_Name'] == 'Victoria Island') & (df_combined['State'] != 'Lagos') & (df_combined['State'].notnull())]
```

```python
# # replace with Lagos
for key, row in df_combined.iterrows():
    if(row.LGA_Name == 'Victoria Island'):
        df_combined['State'].fillna('Lagos', inplace=True)
```

```python
# Gender 
'''
We will have categories Male, Female, and Other. Let's get on with it.
'''
mapper = {
    'Male': 'Male',
    'Female': 'Female',
    'Entity': 'Other',
    'Joint Gender': 'Other',
    'NOT STATED': 'Other',
    'NO GENDER': 'Other',
    np.nan : np.nan,
    'SEX': 'Other'
}

df_combined['Gender'] = df_combined['Gender'].apply(lambda x: mapper[x])

'''
I will then replace the missing values with the most frequent Gender which is Male
'''
df_combined['Gender'].fillna('Male', inplace=True)
```

```python
df_combined['Gender'].value_counts()
```

```python
# Product Name
'''
Most Occurring Product Name is Car Classic. Replace all missing values with Car Classic
'''
df_combined['ProductName'].fillna('Car Classic', inplace=True)
```

```python
claims = df_combined['target'].value_counts()
'''
Most People did not submit a claim. So I will replace missing target values with 0
'''
print('Number of claims NOT  submitted is {} while number of claims submitted is {}.'.format(*df_combined['target'].value_counts()))

# replace missing with 0
df_combined['target'].fillna(0, inplace=True)
```

#### Feature Engineering
- Extract information from other columns in the dataset

```python
date_columns = ['Policy Start Date', 'Policy End Date', 'First Transaction Date']
date_extractions = ['year', 'month', 'day', 'quarter', 'week']
for col in date_columns:
    for feature in date_extractions:
        df_combined[col+"_"+feature] = getattr(df_combined[col].dt, feature)

        
# get weekend 0, 1 (1=weekend, 0=not weekend)
for col in date_columns:
    df_combined[col+"_"+'weekend'] = np.where((df_combined[col].dt.day_name()).isin(['Sunday','Saturday']), 1, 0) 
```

#### Modeling

```python
# Separate training set and test test
training_df = df_combined[:training_data]
testing_df = df_combined[training_data:]

main_cols = [col for col in testing_df.columns if "Target" not in col] # remove target variable in test test
testing_df = testing_df[main_cols]

# training and testing shapes (Original Format)
training_df.shape, testing_df.shape
```

```python
X = training_df.iloc[:, training_df.columns != 'Target']
y = training_df.iloc[:, training_df.columns == 'Target']
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
X, y = sm.fit_resample(X, y)
```

```python
# Split data into training set and test test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
# Feature scale the data
pipeline = Pipeline([
    ('std_scaler', StandardScaler())
])
X_train = pipeline.fit_transform(X_train)
X_test = pipeline.transform(X_test)

# scale provided test set
testing_df = pipeline.fit_transform(testing_df)
```

## **Init Model**

```python
def train_predict(classifier, X_train, y_train, X_test, y_test):
    ''''''
    results = {}
    
    # Training
    start = time()
    clf = classifier.fit(X_train, y_train)
    end = time()
    
    results['train_time'] = end - start
    
    # Testing
    start = time()
    predictions = clf.predict(X_test)
    end = time()
    
    results['test_time'] = end - start
    
    # Accuracy score
    score = accuracy_score(y_test, predictions)
    results['accuracy_score'] = score
    
    # FBeta_score
    fb_score = fbeta_score(y_test, predictions, beta=0.5)
    f_score = f1_score(y_test, predictions)
    results['f1_score(0.5)'] = fb_score
    results['f1_score'] = f_score
    
    print('{} has been trained and tested \U0001F600'.format(classifier.__class__.__name__))
    return results
```

```python
clf_A = RandomForestClassifier()
clf_B = SGDClassifier()
clf_C = LogisticRegression()
clf_D = GradientBoostingClassifier()
clf_E = DecisionTreeClassifier()

results = {}
for i, clf in enumerate([clf_A, clf_B, clf_C, clf_D, clf_E]):
    clf_name = clf.__class__.__name__
    results[clf_name] = {}
    results[clf_name][i] = train_predict(clf, X_train, y_train, X_test, y_test)
    
display(results)
```
